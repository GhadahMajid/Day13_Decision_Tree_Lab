# Q1: Import labraries

```python
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Q2: Read instagram_users.csv file

```python
df = pd.read_csv('instagram_users.csv')
df.head()
```

```python
df.info()
```

# Q3: Split tha dataset into training and testing

```python
X = df.drop('real_fake',axis=1)
y = df['real_fake']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)# 70% - 30%
```

```python
df[1:7]
```

# Q4: Build three machine models 

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from sklearn.metrics import plot_confusion_matrix
```

## Q4.1: The first machine model
- Print the model's name.
- Print the model's accuracy.
- Print the model's confusion matrix.

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)

print("Model's Accuracy:" , type(dtree).__name__)
print("Model's Accuracy:" , accuracy_score(y_test,predictions))
print("Model's Confusion Matrix:")
print(confusion_matrix(y_test,predictions))
plot_confusion_matrix(dtree, X_test, y_test)  
plt.show()
```

## Q4.2: The second machine model
- Print the model's name.
- Print the model's accuracy.
- Print the model's confusion matrix.

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100) 
rfc = rfc.fit(X_train, y_train)
predictions = rfc.predict(X_test)

print("Model's Accuracy:" , type(rfc).__name__)
print("Model's Accuracy:" , accuracy_score(y_test,predictions))
print("Model's Confusion Matrix:")
print(confusion_matrix(y_test,predictions))
plot_confusion_matrix(rfc, X_test, y_test)  
plt.show()
```

## Q4.3: The third machine model
- Print the model's name.
- Print the model's accuracy.
- Print the model's confusion matrix.

```python
from sklearn.linear_model import LogisticRegression

logisticRegr = LogisticRegression()
logisticRegr = logisticRegr.fit(X_train,y_train)
predictions = logisticRegr.predict(X_test)

print("Model's Accuracy:" , type(logisticRegr).__name__)
print("Model's Accuracy:" , accuracy_score(y_test,predictions))
print("Model's Confusion Matrix:")
print(confusion_matrix(y_test,predictions))
plot_confusion_matrix(logisticRegr, X_test, y_test)  
plt.show()
```
